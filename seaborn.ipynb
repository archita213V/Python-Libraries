1. Histogram plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df
```

```python
sns.histplot(df["total_bill"], bins=10)
plt.show()
```

KDE plot

```python
sns.kdeplot(df["total_bill"], shade=False)
plt.show()
```

KDE + Histogram

```python
sns.histplot(df["total_bill"], bins=10, kde=True)
sns.kdeplot(df["total_bill"], shade=False)
plt.show()
```

```python
df['size'].unique()
```

```python
plt.subplot(1,2,1)
sns.histplot(df["total_bill"], bins=10, kde=True)
plt.subplot(1,2,2) 
sns.histplot(df["size"], bins=10, kde=True)
plt.show()
```

joint plot

```python
sns.jointplot(x=df["total_bill"], y=df["tip"], kind="scatter")
```

```python
sns.jointplot(x=df["total_bill"], y=df["tip"], kind="reg")#kdeplot
```

Pair plot

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

2. Categorical plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df
```

Count plot

```python
sns.countplot(x=df['sex'],hue=df['smoker'])
```

Bar plot

```python
sns.barplot(x=df['sex'],y=df['total_bill'],hue=df['smoker'])
```

```python
import numpy as np
sns.barplot(x=df['sex'],y=df['total_bill'],hue=df['smoker'],estimator=np.std)
```

Box plot

```python
sns.boxplot(x=df['day'],y=df['total_bill'],hue=df['smoker'])
plt.show()
```

Violin plot

```python
sns.violinplot(x=df['day'],y=df['total_bill'])
```

strip plot

```python
sns.stripplot(x=df['day'],y=df['total_bill'],hue=df['smoker'])
```

swarm plot

```python
sns.swarmplot(x=df['day'],y=df['total_bill'],hue=df['smoker'])
```

```python
sns.violinplot(x= 'tip',y = 'day',data = df)
sns.swarmplot(x= 'tip',y = 'day',data = df)
```

3. Matrix plot

Heatmap plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips= sns.load_dataset("tips")
tips
```

```python
flights=sns.load_dataset("flights")
flights
```

```python
tipscorr=tips[['total_bill','tip','size']]
tipscorr
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(),annot=True)
```

```python
flightscorr=flights[['year','passengers']]
flightscorr
```

```python
flightscorr.corr()
```

```python
sns.heatmap(flightscorr.corr(),annot=True)
```

Cluster map

```python
sns.clustermap(tipscorr.corr(),annot=True)
```

```python
sns.clustermap(flightscorr.corr(),annot=True)
```

Pivot heat table

```python
pvtflights=flights.pivot_table(index='month',columns='year',values='passengers')
pvtflights
```

```python
sns.heatmap(pvtflights)
```

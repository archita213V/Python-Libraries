```python
a=[1,2,3,4,5,6,7]
b=[10,20,30,40,50,60,70]
plt.plot(a, b)
plt.xlabel('a')
plt.ylabel('b')
plt.title('Line Graph')
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(a,b)
plt.subplot(2,2,2)
plt.bar(a,b)
plt.subplot(2,2,3)
plt.scatter(a,b)
plt.subplot(2,2,4)
plt.hist(b)
plt.tight_layout()
plt.show()
```

object oriented way

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7]
y = [10,20,30,40,50,60,70]

fig = plt.figure(figsize=(10,10))

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x, y)
axes1.set_title("Main Plot")

axes2 = fig.add_axes([0.5, 0.5, 0.3, 0.3])
axes2.plot(y, x)
axes2.set_title("Inset Plot")

plt.show()
```

Pie plot

```python
subjects = ["Math", "Physics", "Chemistry", "English"]
hours = [4, 3, 2, 1]
colors = ["red", "blue", "green", "orange"]
plt.pie(
    hours,
    labels=subjects,
    colors=colors,
    autopct="%1.1f%%"
)

plt.show()
```

Label and legend

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [10,20,30,40]
plt.plot(x, y, label="Sales")
plt.legend()
plt.show()
```

```python
#multiple plots
x = [1,2,3,4]
y1 = [10,20,30,40]
y2 = [15,25,35,45]
plt.plot(x, y1, label="Product A")
plt.plot(x, y2, label="Product B")
plt.legend()
plt.show()
```

Text annotation

```python
y = [543,765,345,876,134,234,567,890,123]
x = range(1,10)
min_y=min(y)
x_co=x[y.index(min_y)]
plt.plot(x, y, label="Sales")
plt.annotate("lowest point",xy=(x_co,min_y), xytext=(x_co+0.5, min_y+100), arrowprops=dict(facecolor='black', shrink=0.05)) 
plt.legend()
plt.show()
```

Working with images

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

Display images

```python
img=mpimg.imread('image.png')
print(img)
```

```python
plt.imshow(img)
plt.show()
```

```python
plt.axis('off')
plt.imshow(img)
plt.show()
```

Image shape

```python
print(img.shape)
```

Convert to grayscale

```python
plt.imshow(img,cmap='gray')
```

Crop images

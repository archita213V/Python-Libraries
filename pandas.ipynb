SERIES 
(one dimensional array ,stores various type of data)

```python
import pandas as pd
#using list
x=[3,4,5,6,7]
var=pd.Series(x)
print(var)
print(type(var))
print(var[0])
print(var[1])

```

```python
#using index
y=[7,58,89,46,97]
var1=pd.Series(y,index=['a','b','c','d','e'],dtype=int)
print(var1)
```

```python
#using dictionary
dic={'name':['archita','Rahul','prince'],'age':[20,23,35],'city':['delhi','mumbai','kolkata']}
var2=pd.Series(dic)
print(var2)
```

```python
print(var+var1)
```

DATAFRAMES (multidimensional array 2d,3d...)

```python
import pandas as pd
#using list
lsit=[1,2,3,4,5]
var=pd.DataFrame(lsit)
print(var)
print(type(var))
```

```python
#using dictionary
dic={'name':['archita','Rahul','prince'],'age':[20,23,35],'city':['delhi','mumbai','kolkata']}
var2=pd.DataFrame(dic)
print(var2)
```

```python
#using index
dic={'name':['archita','Rahul','prince'],'age':[20,23,35],'city':['delhi','mumbai','kolkata']}
var2=pd.DataFrame(dic,columns=['name','age'],index=[0,1,2])
print(var2)
```

```python
#using series
sr={'s':pd.Series([1,2,3,4,5]),'r':pd.Series([6,7,8,9,10])}
var3=pd.DataFrame(sr)
print(var3)
```

ARITHMETIC OPERATION (+,-,/,*,%,**)

```python
import pandas as pd
var=pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})
print(var)
#operations
var['C']=var['A']+var['B']
var['D']=var['A']-var['B']
var['E']=var['A']*var['B']
var['F']=var['A']/var['B']  
print(var)
#conditions
var['X']=var['A']<=15
var['Y']=var['B']>=16
print(var)
```

INSERT AND DELETE FUNCTIONS IN PANDAS

INSERT

```python
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}
df = pd.DataFrame(data)
print(df)
df.insert(2, 'Country', ['USA', 'USA', 'USA', 'USA'])
print(df)
```

DELETE

```python
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}
df = pd.DataFrame(data)
print(df)   
var=df.pop('City')
print(var)
print(df)
del df['Age']
print(df)
```

MERGE AND CONCAT DATA

MERGE

```python
import pandas as pd
var1=pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})
var2=pd.DataFrame({'A':[1,2,3],'B':[14,15,16]})
pd.merge(var1,var2,on='A')
```

```python
var1=pd.DataFrame({'A':[1,2,4],'B':[4,5,6]})
var2=pd.DataFrame({'A':[1,2,3],'B':[14,15,16]})
pd.merge(var1,var2,on='A',how='inner')# by default inner join is used
```

```python
var1=pd.DataFrame({'A':[1,2,4],'B':[4,5,6]})
var2=pd.DataFrame({'A':[1,2,3],'B':[14,15,16]})
pd.merge(var1,var2,on='A',how='left')#or right join can also be used
# how='outer' can also be used for full outer join
#indicater=True can also be used to show the source of each row in the merged dataframe
```

```python
var1=pd.DataFrame({'A':[1,2,4],'B':[4,5,6]})
var2=pd.DataFrame({'A':[1,2,3],'B':[14,15,16]})
pd.merge(var1,var2,left_index=True, right_index=True,suffixes=('_left','_right'))
```

```python
import pandas as pd
d1=pd.DataFrame({'A':[1,2,4],'B':[4,5,6]})
d2=pd.DataFrame({'A':[1,2,3],'B':[14,15,16]})
pd.concat([d1,d2],axis=0)#row
pd.concat([d1,d2],axis=1)#column
```
